```python
import pandas as pd
import glob
```

```python
file_list =  glob.glob("./result/*.csv")
len(file_list)
```

```python
for file in file_list:
    df = pd.read_csv(file)
    if len(df) > 0:
        name = df.iloc[0, 1]
        print(name)
        df = df[df["title"].str.contains(name, na=False)]
        file_path = file.split('\\')[-1]
        df.to_csv(f"./result_temp/{file_path}", index=False, encoding='utf-8-sig')
```
